```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataf = pd.read_csv('Housing.csv')
print(dataf)
```

```python
plt.scatter(dataf['area'], dataf['price'], color='red', alpha=0.5)
plt.title('Price vs Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(dataf['price'], bins=37, kde=True, color='red')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10,6))
plt.title('Price by Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
pal = ['g', 'b', 'r', 'y']
sns.boxplot(x='bathrooms', y='price', data=dataf, palette=pal)
plt.ylabel('Price')
```

```python
plt.figure(figsize=(12,5))
plt.title('No. of Parkings by Area')
plt.xlabel('Number of Parkings')
sns.boxplot(x='parking', y='area', data=dataf, palette='deep')
plt.ylabel('Area in sq.ft.')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='mainroad', y='price', data=dataf)
plt.title('Main Road Access vs Price')
plt.xlabel('Main Road Access (Yes/No)')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='bedrooms', y='price', data=dataf)
plt.title('Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(9,10))
sns.barplot(x='furnishingstatus', y='price', data=dataf, palette='deep',)
plt.title('Furnishing Status vs Price')
plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.show()
```
